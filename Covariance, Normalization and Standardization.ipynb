# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
s = pd.Series([1000,2000,3000,5000,9000])
s      
```

```python
s.min()
```

```python
s.max()
```

```python
s_norm = (s-s.min()) / (s.max() - s.min())
print(s_norm)
```

```python
s_norm = MinMaxScaler().fit_transform(s.values.reshape(-1,1))
print(s_norm)
```

```python
df = pd.DataFrame({
    "speed": [40,60,80,100,120],
    "distance": [768,564,456,345,250]
})
df
```

```python
df.min()
```

```python
df.max()
```

```python
norm_data =  MinMaxScaler().fit_transform(df)
norm_data
```

```python
norm_df = pd.DataFrame(norm_data, columns=df.columns)
norm_df
```

# Standardization

```python
s = pd.Series([1000,2000,3000,5000,9000])
s      
```

```python
s.mean()
```

```python
s.std()
```

```python
Z_Score=(s - s.mean())/s.std()
print(Z_Score)
```

```python
from sklearn.preprocessing import StandardScaler
Z_Score = StandardScaler().fit_transform(s.values.reshape(-1,1))
print(Z_Score)
```

```python
Z_Score=(s - s.mean())/s.std(ddof=0)
print(Z_Score)
```

```python
df = pd.DataFrame({
    "speed": [40,60,80,100,120],
    "distance": [768,564,456,345,250]
})
df
```

```python
Z_Score = StandardScaler().fit_transform(df)
Z_Score_df = pd.DataFrame(Z_Score, columns=df.columns)
print(Z_Score_df)
```
